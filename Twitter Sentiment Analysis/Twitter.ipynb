```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# ps=PorterStemmer()
```

```python
def tweetsToWords(tweet):
    letters = re.sub('[^a-zA-Z]',' ',tweet)
    words = letters.lower().split()
    stops = set(stopwords.words('english'))
    usefulWords = [w for w in words if w not in stops]
    return ' '.join(usefulWords)
```

```python
tweet_df_train['clean_tweet'] = tweet_df_train['text'].apply(lambda x: tweetsToWords(x))
```
